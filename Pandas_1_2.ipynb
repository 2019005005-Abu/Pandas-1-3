```python
##Combines 2 pandas series
data1=[1,2,3,4,5,6,7]
data2=[8,9,10,11,12,13,14]
data3=[15,16,17,18,19,20,21]
df1=pd.DataFrame(data1);
df2=pd.DataFrame(data2);
df3=pd.DataFrame(data3);

series1=pd.Series(data1);
series2=pd.Series(data2);
series3=pd.Series(data3);
def demo(x1,x2):
    if x1>x2:
        return x1;
    else:
        return x2;
result1=series1.combine(series2,demo);
result2=series2.combine(series3,demo);
result3=series3.combine(series1,demo);

print(f"result:1:{result1}")
print(f"result:2:{result2}")
print(f"resul3:3:{result3}")

```

```python
'''Pandas-categorical series'''
'''
1.Append new category
2.Remove a category
'''
s=pd.Series(['a','b','c','d','e','f'],dtype="category")
print(s);
s=s.cat.add_categories('t')
print(s)
s=s.cat.add_categories('m');
print(s);

re=s.cat.remove_categories('t');
print(re);

```
